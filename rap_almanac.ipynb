```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
file = open("rap_almanac_scrape_1979-2018.txt")
soup = BeautifulSoup(file, 'html.parser')
```

```python
rows = soup.find_all("tr")
```

```python
df = pd.DataFrame(columns=["Title", "Album", "Artist", "Release_Date", "City", "State", "County", "Country", "Lyrics"])
```

```python
df
```

```python
list(filter(None, rows[0].get_text().split("\n")))
```

```python
data=[]
for row in rows:
    data.append(list(filter(None, row.get_text().split("\n")))[0:9])
```

```python
df=pd.DataFrame(data, columns=["Title", "Album", "Artist", "Release_Date", "City", "State", "County", "Country", "Lyrics"])
```
